```python
import csv
f = open('Voting_Poll_Sites_Map.csv','r')
reader = csv.reader(f)

next(reader)

postcode_count = {}
# count = 0
for row in reader:
    code = row[7]
    #print(code)
#code=key
#count=value
    if code in postcode_count.keys():
        postcode_count[code]+=1
    else:
        postcode_count[code]=1
print(postcode_count)
```

```python
import csv
f = open('Demographic_Statistics_By_Zip_Code.csv','r')
reader = csv.reader(f)

import pandas as pd
df = pd.read_csv('Demographic_Statistics_By_Zip_Code.csv')
df
# reduced = df[['POSTCODE', 'count_participants', 'count_female', 'percent_female', 'count_male', 'percent_male', 'count_hispanic_latino', 'percent_hispanic_latino', 'count_white_non_hispanic', 'percent_white_non_hispanic', 'count_black_non_hispanic', 'percent_black_non_hispanic','count_receives_public_assistance', 'percent_receives_public_assistance', 'count_nreceives_public_assistance', 'percent_nreceives_public_assistance', 'count_public_assistance_unknown', 'percent_public_assistance_unknown','count_public_assistance_total', 'percent_public_assistance_total']]
# reduced

reduced = df[['POSTCODE','count_participants', 'count_female', 'percent_female', 'count_male', 'percent_male', 'count_white_non_hispanic', 'percent_white_non_hispanic', 'count_black_non_hispanic', 'percent_black_non_hispanic']]
# reduced
# print(postcode_count['10001'])

def myfunction(row):
    pc = str(row['POSTCODE']).split('.')[0]
    # print(type(pc))
    # print(pc)
    if pc not in postcode_count.keys():
        return 0
    return postcode_count[pc] 

reduced['pc_count'] = 0

reduced['pc_count'] = df.apply(myfunction, axis=1)
#reduced

def myfunction2(row):
    pc = str(row['POSTCODE']).split('.')[0]
    # print(type(pc))
    # print(pc)
    if pc not in sq_area.keys():
        return None
    return sq_area[pc]

reduced['sq_area'] = 0

reduced['sq_area'] = df.apply(myfunction2, axis=1)
reduced
```
